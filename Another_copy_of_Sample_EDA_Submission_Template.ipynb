#### Chart - 10

```python
# Chart - 10 visualization code
 # Ensure 'country_txt' and 'INT_ANY' are valid
df = df[df['country_txt'].notna() & df['INT_ANY'].notna()]
df['INT_ANY'] = df['INT_ANY'].astype(int)

# Group by country and INT_ANY (0 or 1) and count incidents
country_int_counts = df.groupby(['country_txt', 'INT_ANY']).size().unstack(fill_value=0)

# Get top 10 countries by total number of incidents
top_countries = country_int_counts.sum(axis=1).sort_values(ascending=False).head(10)

# Subset data
top_countries_data = country_int_counts.loc[top_countries.index]

# Plotting
top_countries_data.rename(columns={0: 'Domestic', 1: 'International'}, inplace=True)
top_countries_data.plot(kind='barh', stacked=True, figsize=(12, 7), colormap='coolwarm')

plt.title('Domestic vs International Terrorist Attacks (Top 10 Countries)', fontsize=14)
plt.xlabel('Number of Incidents')
plt.ylabel('Country')
plt.legend(title='Attack Type')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
